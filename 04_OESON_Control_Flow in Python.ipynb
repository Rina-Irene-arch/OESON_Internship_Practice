# Control Flow in Phyton
1. Introduction to control flow
Control flow in programming refers to the order in which individual statements, instructions, or functions calls are executed or evaluated. Understanding control flow is essential to solve the complex problems.
2. Conditional statements
Conditiobal statements allow you to execute certain pieces of code based on where conditioln is true or false

## If Statement
The if statements executes a block of code if a specified condition is true

```python
x = 10
if x > 5:
 print('x is larger than 5')
```

## If-else Statement
The if-else statement executes one block of code if the condition is true and another block if the condition is false.

```python
x = 3
if x > 5:
 print('x is greater than 5')
else:
 print('x is not greater than 5')
```

```python
## 'if-elif-else' Statement

x = 10
if x > 15:
 print('x is greater than 15')
elif x > 5:
 print('x is greater than 5 but less than or equal to 15')
else:
 print('x is 5 or less')
```

## Looping statement
Looping statement allow you execute a block of code repeadtely.

### 'for' loop
The for loop itterates over the sequence (such as alist, tuple, dictionary, set, or string)

```python
for i in range(5):
    print(i)
```

### while Loop
The while loop ececutes a block of code as long as a specified condition is true

```python
count = 0
while count < 5:
    print(count)
    count +=1
```

# Control Flow Tools
Phyton  provides several tools to control the flow of loops and conditionals.

#### break statement
the break statement terminates are the loop prematuraly

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

### continue statement
The continue statement skips the rest of the code inside the loop for the current itteration and moves to the next iteration

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

### pass statement
The pass statemnet does nothing and it's used as placeholder

```python
for i in range(10):
    if i % 2 == 0:
        pass
    else:
        print(i)
```

```python
# Wtite the program to print all prime numbers berween 1 and 100

# A prime number is a natural number greater than 1
# that has no positive divisors other than 1 and itself.
# if num is not a prime number,
# it must have a divisor less than or equal to its square root.
# This is a fundamental principle in number theory
# that helps optimize algorithms for finding prime numbers.
for num in range (2, 101):
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num)
```

```python
# Write the program to find the sum of digits of a number
number = 12345
total = 0
while number > 0:
    # get the last digit 12345 % 10 = 5
    digit = number % 10
    total += digit
    # removes the last digit 12345//10 = 1234
    number //= 10
print("Sum of digits:", total)
```

```python
# Write a program to check if a given string is a palindrome (reads the same backward as forward).
s = "radar"
# uses Python's slicing feature to reverse the string
is_palindrome = s == s[::-1] # checks if the string s is equal to its reverse.
print("Is palindrome:", is_palindrome)
```

In Python, the slicing syntax [start:stop:step] allows you to extract parts of a string (or list) by specifying where to start the slice, where to end, and the step, which indicates the intervals at which elements are included in the slice.

For the special case of reversing a string, you use [::-1]. Here's what each part represents:

start: By default, it's set to None, meaning Python will consider the entire sequence from the beginning.
stop: Also defaults to None, meaning it will go until the end of the sequence.
step: When set to -1, it tells Python to go one step backward at a time, hence reversing the order of elements.

```python
# Write a program to count the number of vowels and consonants in a string.
s = "hello world"
vowels = "aeiou"
vowel_count = 0
consonant_count = 0
for char in s.lower():
#     Iterates over each character in the string after converting it to lowercase
#     to ensure the comparison is case-insensitive.
     if char in vowels:
        vowel_count += 1
#         checks if the character is in the vowels string.
#         If true, it increments the vowel_count by 1.
     elif char.isalpha():
#            If the character is not a vowel but is an alphabetic character (ignores spaces, numbers, punctuation, etc.),
#            it increments the consonant_count by 1.
        consonant_count += 1
print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
```

```python
# Write a program to transpose a given matrix.
matrix = [
     [1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]
]
transposed_matrix = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
# range(len(matrix[0])) -      generates a sequence of indices from 0 to the length of the first row in the matrix (in this case, 0, 1, 2).
#                              This represents the columns of the original matrix.
# [row[i] for row in matrix] - The inner list iterates over each row in the matrix and retrieves the element at the index i.
#                              This effectively collects the elements from each row at the same column index.
# [[row[i] for row in matrix] for i in range(len(matrix[0]))] - creates a new list for each index i,
#                              where each element in that list is the result of the inner list comprehension.
print("Transposed matrix:", transposed_matrix)
```
